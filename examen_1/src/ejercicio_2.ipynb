## Ejercicio 2

- Vector inicial: Genera un vector **x** con valores igualmente espaciados entre -10 y 10
- Genera funciones para: f_1(x) = x^2 y f_2(x) = x^3
- Crear un data frame con 3 columnas x, f1(x)m f2(x)
- Grafica las figuras de las funciones usando los datos del data frame

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return x**2
```

```python
def f2(x):
    return x**3
```

```python
# Generamos un vector x usando list comprenhensions
x = [i * (1/2) for i in range (-20, 21)]
```

```python
f1_vals = [ f1(i) for i in x ]
f2_vals = [ f2(i) for i in x ]
```

```python
f1_vals
```

```python
f2_vals
```

```python
df = pd.DataFrame({"x": x, "f1(x)": f1_vals, "f2(x)": f2_vals})
df
```

```python
plt.plot(x, f1_vals, label="f1(x)")
plt.legend()
plt.plot(x, f2_vals, label="f2(x)")
plt.legend()
plt.grid()
plt.xlabel("Vector x")
plt.ylabel("Funciones")
plt.title("Graficación de funciones cuadraticas y cubicas")
```
